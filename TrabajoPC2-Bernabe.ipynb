```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Camilo0702/PC2-Bernabe/main/File-data/HDI%20(1).xlsx'
data = pd.read_excel(url, engine='openpyxl', skiprows = 4)
borrar = data.columns[[3,5,7,9,11,13]]
data = data.drop(borrar, axis = 1)
data.columns = data.columns.str.replace('Unnamed.*','', regex=True)
data = data.round(2)
data.rename(columns={'Life expectancy at birth': 'Life exp at birth', 'Expected years of schooling': 'Exp years of schooling', 'Mean years of schooling': 'Mean years'}, inplace=True)
data['Life exp at birth'] = pd.to_numeric(data['Life exp at birth'], errors='coerce')
data['Exp years of schooling'] = pd.to_numeric(data['Exp years of schooling'], errors='coerce')
data['Mean years'] = pd.to_numeric(data['Mean years'], errors='coerce')
data['Gross national income (GNI) per capita'] = pd.to_numeric(data['Gross national income (GNI) per capita'], errors='coerce')
data['Life exp at birth'] = data['Life exp at birth'].round(2)
data['Exp years of schooling'] = data['Exp years of schooling'].round(2)
data['Mean years'] = data['Mean years'].round(2)
data['Gross national income (GNI) per capita'] = data['Gross national income (GNI) per capita'].round(0)
data = data.astype(object)
data.fillna('', inplace=True)
data
```
